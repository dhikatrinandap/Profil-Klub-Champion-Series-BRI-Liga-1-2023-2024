# **Data Cleaning & Understanding**
Sebelum masuk ke dalam analisis, kita perlu mengenal dataset kita lebih jauh dalam tahapan data understanding. Dari proses ini, kita akan tahu anomali-anomali apa saja yang terdapat di dalam dataset kita dan perlu ditangani dalam tahapan data cleaning. Setiap penangan anomali yang dilakukan, akan disertai dengan justifikasi langkah yang diambil, baik secara domain knowledge maupun secara statistik.

```python
# Mengetahui missing value pada kolom 'market_value'
df[df['market_value'].isna()]
```

```python
# Mengisi missing value pada kolom 'market_value' dengan '0'
df['market_value'].fillna(0, inplace=True)
```

Dikarenakan tidak mendapatkan informasi setelah mencari refenrensi-referensi mengenai market value dari ke-3 pemarin tersebut maka akan diisi 0.

```python
df.isna().sum()
```

```python
# Mengubah format tinggi badan dari string ke float
df['height'] = df['height'].str.replace(',', '.').str.replace('m', '').astype(float)
```

```python
# Mengubah format kolom joined ke datetime
df['joined'] = pd.to_datetime(df['joined'], dayfirst=True, errors='coerce')
```

```python
# Membuat Kolom 'birth_date' dari Kolom 'Date of birth / Age'
df["birth_date"] = df["Date of birth / Age"].str[:-5] 
```

```python
# Peta nama bulan dari bahasa Indonesia ke bahasa Inggris
bulan_indo_to_eng = {
    'Jan': 'Jan', 'Feb': 'Feb', 'Mar': 'Mar', 'Apr': 'Apr', 'Mei': 'May', 'Jun': 'Jun',
    'Jul': 'Jul', 'Agt': 'Aug', 'Sep': 'Sep', 'Okt': 'Oct', 'Nov': 'Nov', 'Des': 'Dec'
}

# Mengonversi nama bulan dari bahasa Indonesia ke bahasa Inggris di kolom 'TL'
for indo, eng in bulan_indo_to_eng.items():
    df['birth_date'] = df['birth_date'].str.replace(indo, eng)

# Membersihkan data dari spasi ekstra atau karakter yang tidak diinginkan
df['birth_date'] = df['birth_date'].str.strip()

# Mengonversi kolom 'birth_date' ke format datetime
df['birth_date'] = pd.to_datetime(df['birth_date'], format='%d %b %Y', dayfirst=True, errors='coerce')
```

```python
# Tanggal referensi
reference_date = pd.to_datetime('2024-05-31')

# Menghitung umur dari tanggal lahir
df['age'] = (reference_date - pd.to_datetime(df['birth_date'])).dt.days // 365
```

```python
# Menhapus kolom 'contract' dan 'Date of birth/Age'
df = df.drop(['contract', 'Date of birth / Age'], axis=1)
```

### **4. Jumlah Pemain Berdasarakan Posisi Tiap Club/Tim**

```python
n_posisi_club = df.groupby(['position', 'club_name'])[['player_name']].count().reset_index().sort_values('player_name', ascending=False)
n_posisi_club.head()
```

```python
# Membuat layout 2x2 untuk 4 grafik
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(25, 10))

# Iterasi untuk setiap klub dan membuat grafiknya
for ax, club in zip(axes.flatten(), n_posisi_club['club_name'].unique()):
    club_data = n_posisi_club[n_posisi_club['club_name'] == club]
    club_data = club_data.sort_values(by='player_name', ascending=True)  # Mengurutkan nilai pasar menurun
    ax.barh(club_data['position'], club_data['player_name'], color='steelblue')
    ax.set_xlabel('Jumlah Pemain')
    ax.set_ylabel('Posisi')
    ax.set_title(f'Jumlah Pemain Berdasarakan Posisi {club}')
```
